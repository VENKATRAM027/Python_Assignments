```python
import time

def concat_with_plus(strings):
    result = ''
    for s in strings:
        result += s
    return result

def concat_with_join(strings):
    return ''.join(strings)

def main():
    # Create a large number of strings to concatenate
    num_strings = 10000
    strings = [f"string{i}" for i in range(num_strings)]

    # Time the + operator approach
    start_time = time.time()
    concat_with_plus(strings)
    plus_duration = time.time() - start_time
    print(f"Time taken with + operator: {plus_duration:.5f} seconds")

    # Time the join approach
    start_time = time.time()
    concat_with_join(strings)
    join_duration = time.time() - start_time
    print(f"Time taken with join method: {join_duration:.5f} seconds")

if __name__ == "__main__":
    main()
```

```python
# Write a function that counts the occurrences of each character in a string and returns a 
# dictionary with the results. 

def count_characters(s):
    char_count = {}

    for char in s:
        if char in char_count:
            char_count[char] += 1

        else:
            char_count[char] = 1

    return char_count

input_string = 'Python is Awesome!'
result = count_characters(input_string)
print(result)        
```

```python
# Create a function that checks if a string is a palindrome (reads the same backward as 
# forward), ignoring case, spaces, and punctuation.

def is_palindrome(data):
    if data.lower() == data[::-1].lower():
        return True
    else:
        return False
    
    
data = 'racecar'
result = is_palindrome(data)
print(result)    
```
