```python
# Find the maximum and minimum value from a list without using max/min.

l2 = [2,.3,400,55,44,77,8,9,99,100]

min_value = l2[0]
max_value = l2[0]


for i in l2:
    if i < min_value:
        min_value = i
    if i > max_value:
        max_value = i

print('Minimum value:',min_value)
print('Maximmum value:',max_value)    

```

```python
#  Merge two lists and remove duplicates. 

def merge_and_remove_duplicates(list1,list2):
    combined_list = list1 + list2
    return list(set(combined_list))

list1 = [1,3,5,7,9,3,5,7,9]
list2 = [2,4,6,8,10,2,4,6,8]
print(merge_and_remove_duplicates(list1,list2))
```

```python
#  Write a function to check if a list is a palindrome. 

def check_palindrome(list4):
    if list4 == list4[::-1]:
        return 'The given list is a palindromic list'
    else:
        return 'It is not a palindromic list'
    

list4 = [1,2,3,2,1]
result = check_palindrome(list4)
print(result)    
```

```python
#  Create a deep copy of a nested list and prove the difference with an example. 

import copy

original_list = [[1,2,3],[4,5,6]]

deep_copy_list = copy.deepcopy(original_list)

shallow_copy_list = copy.copy(original_list)

original_list[0][0] = 'Modified'

print('Original list:',original_list)
print('Deep copy:',deep_copy_list)
print('Shallow copy:',shallow_copy_list)
```

```python
#  Show the effect of modifying a shallow copied list on the original. 

og_list = [1,2,3,4,5]

shallow_copy = copy.copy(og_list)
shallow_copy[1] = 99

print(shallow_copy)
```

```python
# Write a function that returns a tuple of two values. 

def sum_and_product(x,y):
    sum_value = x + y
    product_value = x * y
    return sum_value,product_value

result = sum_and_product(8,4)
print(result)
```

```python
# Print the memory size of a tuple vs list of same elements.

import sys

list1 = [1,2,3,4,5]
tuple1 = (1,2,3,4,5)

size_of_list = sys.getsizeof(list1)
size_of_tuple = sys.getsizeof(tuple1)

print(f"The memory size of list:",size_of_list)
print(f"The memory size of tuple:",size_of_tuple)
```

```python
#  Use _asdict() method on a namedtuple and display keys.

from collections import namedtuple

from collections import namedtuple

# 1. Define the namedtuple type
# Let's create a 'Book' structure
Book = namedtuple('Book', ['title', 'author', 'year', 'isbn'])

# 2. Create an instance of the namedtuple
my_book = Book(
    title="The Great Gatsby",
    author="F. Scott Fitzgerald",
    year=1925,
    isbn="978-0743273565"
)

# 3. Use _asdict() to convert to a dictionary
# 💡 Note: In Python 3.8+, this returns a standard dict
book_dict = my_book._asdict()

# Display the converted dictionary
print(f"📚 Converted Dictionary:\n{book_dict}\n")

# 4. Display the keys
print("🔑 Keys from the dictionary:")
keys = book_dict.keys()

# Method A: Direct display
print(f"Raw Keys View: {keys}")

# Method B: Iterating through keys
print("\nIterated Keys:")
for key in keys:
    print(f"- {key}")

# Method C: As a list
print(f"\nList of Keys: {list(keys)}")
```
