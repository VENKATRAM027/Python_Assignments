```python
#  Make a calculator that returns result only if all inputs are truthy; else returns 'Invalid 
# Input'.

def truthy_calculator(a,b,operator):

    if a and b and operator:
        if operator == '+':
            return a+b
        elif operator == '-':
            return a-b
        elif operator == '*':
            return a*b
        elif operator == '/':
            if b != 0:
                return a/b
            else:
                return 'Invalid Output: Division by Zero'
        else:
            return 'Invalid Input: Unsupported Operation'
    else:
        return 'Invalid Input'        
    
truthy_calculator(100,20,'/')    
```

```python
# Assigning boolean values
A = True
B = False
C = True

# Print the initial values
print(f"A = {A}")
print(f"B = {B}")
print(f"C = {C}")

# Step 1: Evaluate (A and B)
step1 = A and B
print(f"Step 1: Evaluating (A and B) = {A} and {B} = {step1}")

# Step 2: Evaluate (not C)
step2 = not C
print(f"Step 2: Evaluating (not C) = not {C} = {step2}")

# Step 3: Combine results of Step 1 and Step 2 using 'or'
final_expression = step1 or step2
print(f"Step 3: Evaluating ((A and B) or (not C)) = {step1} or {step2} = {final_expression}")

# Final output
print(f"The final expression evaluates to: {final_expression}")
```
