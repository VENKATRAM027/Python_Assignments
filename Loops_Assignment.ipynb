```python
# Write a program to print the factorial of a number using while loop. 

def factorial(n):
    if n < 0:
        return 'Factorial is not defined for negative numbers'
    elif n == 0:
        return 1
    else:
        result = 1
        count = 1

        while count <= n:
            result = result * count
            count += 1

        return result


print(factorial(5))
```

```python
#  Sort a dictionary by its values and print key-value pairs. 

data = {
    'apple' : 150,
    'banana' : 80,
    'guava' : 50,
    'orange' : 70,
    'grapes' : 40
}

sorted_scores = sorted(data.items(), key=lambda item:item[1])

print("sorted scores:")
for fruit,rate in sorted_scores:
    print(f"{fruit} : {rate}")
```

```python
#  Flatten a nested list using nested loops.

nested_list = [[1,2,3],[4,5],[6,7,8,9]]

flattened_list = []

for sublist in nested_list:
    for item in sublist:
        flattened_list.append(item)

print('flattened_list:',flattened_list)        
```

```python
#  Find maximum and minimum in a list using loop (not built-in max/min).

numbers = [55,12,67,75,24,2,99,1000,2550]

min_value = numbers[0]
max_value = numbers[0]

for number in numbers:
    if number < min_value:
        min_value = number
    if number > max_value:
        max_value = number


print("Maximum value :",max_value)
print("Minimum value:",min_value)             
```

```python
# Print all words from a string that start with vowel using loop idiom. 

text = 'An apple a day keeps the doctor away from an umbrella'

vowels = 'aeioeAEIOU'

words = text.split()

print("word starts with vowels:")
for word in words:
    if word[0] in vowels:
        print(word)
```

```python
# Print a square pattern of '*' of size 4x4:

n = 4

for i in range(n):
    for j in range(n):
        print('*',end= '')
    print()    
```

```python
#  Print a right-angled triangle pattern: 

n = 5

for i in range(1,n+1):
    for j in range(i):
        print('*',end = ' ')
    print()    
```

```python
# Print a pyramid pattern: 

n = 5

for i in range(n):
    for j in range(n-i-1):
        print(' ',end = ' ')

    for k in range(2 * i + 1):
        print('*',end = ' ')
    print()         
```
